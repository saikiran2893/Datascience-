```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## importing the dataset
ds = pd.read_csv('sms.csv')

ds.head()
```

```python
##cleaning the dataset
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords ## for removing stop words
from nltk.stem.porter import PorterStemmer ## for stemming the 

corpus = []

for i in range(0,5560):
##removing other characters except alphabets
    mail = re.sub('[^a-zA-Z]',' ',ds['text'][i])
    print(mail)

    ##converting into lower
    mail = mail.lower()
    print(mail)

    ##split the sentences into each text
    mail=mail.split()
    print(mail)

    ps = PorterStemmer()

    mail= [ps.stem(word) for word in mail if not word in set(stopwords.words('english'))]
    mail = ' '.join(mail)

    corpus.append(mail)
```

```python
## converting strings to factors
category = {'ham':1,'spam':0}
ds['type'] = ds['type'].map(category)
ds.info()
ds
```

```python
print(ds.shape)
```

```python
## another moethod to clean the text
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X=cv.fit_transform(corpus).toarray()
Y=ds.iloc[:,0].values
```

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y, test_size=0.30,random_state=0)
```

```python
import sklearn as sk
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(Xtrain,Ytrain)
```

```python
ypred = classifier.predict(Xtest)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Ytest,ypred)
cm
```
